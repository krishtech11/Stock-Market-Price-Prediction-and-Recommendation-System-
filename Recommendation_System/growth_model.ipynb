```python
#importing necessary libraries

import pandas as pd
```

```python
#loading the Dataset

file_path = 'C:\\Users\\krishna\\Downloads\\Recommendation'
stock_data = pd.read_excel("datasets/Recommendation.xlsx")
```

```python
#selecting relevant columns for recommendation

data = stock_data[['Company Name','Symbol','Last Traded Price','365 Day Percentage Change']]
```

```python
#checking the datatypes of each columns

print(data.dtypes)
```

```python
#converting the datatype of percentage change to numeric type as it contains both numeric and string type

data['365 Day Percentage Change'] = pd.to_numeric(data['365 Day Percentage Change'], errors='coerce')
```

```python
#filtering companies with missing percentage change data (Handling NaN values)

data = data.dropna(subset=['365 Day Percentage Change'])
```

```python
#sorting the data by percentage change to get the top growing stocks

top_growing_stocks = data.sort_values(by='365 Day Percentage Change', ascending=False).head(3)
```

```python
#provides the names of top 3 growing companies ,
#their last traded prices and the percentage and 
#change in their stocks over the past year


print("Top 3 Growing Stocks:")
print(top_growing_stocks[['Company Name', 'Last Traded Price', '365 Day Percentage Change']])
```
